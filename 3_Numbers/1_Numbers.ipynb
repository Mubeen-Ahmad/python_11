# Built-in Function for integers

```python
# int --> convert in integer

x = "10"
print(type(x))

x = int(x)
print(type(x))
```

```python
# bin --> convert in binary

x = 10
print(type(x),x)

x = bin(x)
print(type(x),x)
# binary return str data type
```

```python
# hex --> convert in hexadecimal

x = 10
print(type(x),x)

x = hex(x)
print(type(x),x)
# hexadecimal return str data type
```

```python
# oct --> convert in octal

x = 10
print(type(x),x)

x = oct(x)
print(type(x),x)
# octal return str data type
```

# Float Nan Number

<h2><p>NaN standing for not a number, is a numeric data type used to represent any value that is undefined or unpresentable.<br><br>

For example, 0/0 is undefined as a real number and is, therefore, represented by NaN.<br><br>
The square root of a negative number is an imaginary number that cannot be represented as a real number, so, it is represented by NaN.

NaN is also assigned to variables, in a computation, that do not have values and have yet to be computed.<br><br>
Note that the “NaN” passed to the float is not case sensitive. All of the 4 variables come out as nan.
</h2>

```python
n1 = float("nan")
n2 = float("Nan")
n3 = float("NaN")
n4 = float("NAN")

print(n1, n2, n3, n4)
```

# Complex Number
<h3>
Complex numbers are created from two real numbers.<br><br>It is written in the form of (x + yj) where x and y are real numbers <br><br>and j is an imaginary number which is the square root of -1.</h3>

<h3>
<i>real:</i> <br><br>It is a required input and it denotes the real part of the complex number. <br>By default, it is 0. It can also be represented as a string like this ‘1+1j’ and in that case, the second part will be omitted.

imag: It is an optional part and it denotes the imaginary part of the complex number. By default, it is 0.

</h3>

```python
x = 1+1j
print(x,type(x))
```

```python
x = 1-1j
print(x,type(x))
```

```python
x = 1j
print(x,type(x))
```

# Type Conversion 
<h3>

The process of converting the value of one data type (integer, string, float, etc.) to another data type is called type conversion.<br><br>Python has two types of type conversion.

    Implicit Type Conversion
    Explicit Type Conversion

</h3>

# Implicit Type Conversion

<h3>
In Implicit type conversion, Python automatically converts one data type to another data type. This process doesn't need any user involvement.<br><br>Python promotes the conversion of the lower data type (integer) to the higher data type (float) to avoid data loss
</h3>

```python
# int + float = float

10 + 10.1
```

```python
int_ = 10
float_ = 50.5
```

```python
print(int_+float_)
```

# but Implicit Type Conversion are not work with other data type

```python
"10"+10.5
```

# Explicit Type Conversion

<h3>
In Explicit Type Conversion, users convert the data type of an object to required data type. <br><br>We use the predefined functions like int(), float(), str(), etc to perform explicit type conversion.

This type of conversion is also called typecasting because the user casts (changes) the data type of the objects.
</h3>

```python
int('10')+10.5
```

```python
x = "50"
y = "10"

print(x+y)
```

```python
print(int(x)+int(y))
```

```python
# int to bool
x = 1
bool(x)
```

```python
x = 1
x = bool(x)
x+x
```

```python
x = bool(10)
int(x)

# note 1 means true 0 means false only 2 digits in bool
```
