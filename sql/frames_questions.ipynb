<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">Find the branch toppers</div><br>





```python
%%sql

SELECT name,branch,marks  FROM (
    
    SELECT *,
    
        FIRST_VALUE(name) OVER(
            PARTITION BY branch
            ORDER BY marks DESC
        ) AS 'topper_name',
    
        FIRST_VALUE(marks) OVER(
            PARTITION BY branch
            ORDER BY marks DESC
        ) AS 'topper_marks'
        
    FROM 
    marks
    ) t1

WHERE t1.name = t1.topper_name AND t1.marks = t1.topper_marks
```

<br><br>
<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">Find the branch lowers</div><br>





```python
%%sql

SELECT name,branch,marks  FROM (
    SELECT *,
    
    LAST_VALUE(name) OVER lower_n AS `lower_name`,
    LAST_VALUE(marks) OVER lower_n AS `lower_marks`
    
    FROM 
    marks
    
    WINDOW lower_n AS (
                PARTITION BY branch
                ORDER BY marks DESC
                ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING
                )

) t1

WHERE t1.name = t1.lower_name AND t1.marks = t1.lower_marks
```

<div style="text-align:justify; color:black; font-size:20px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">Find the month of month revenue growth</div><br>





```python
%%sql
SELECT 
MONTHNAME(date) AS `month`,
SUM(amount) AS `total` FROM orders
GROUP BY month
ORDER BY date ASC
```

```python
%%sql
SELECT 
MONTHNAME(date) AS `month`,
SUM(amount) AS `total`,
LAG(SUM(amount)) OVER(ORDER BY date) AS `lag`,

SUM(amount) - LAG(SUM(amount)) OVER(ORDER BY date) AS subtract,
((SUM(amount) - LAG(SUM(amount)) OVER(ORDER BY date)) / LAG(SUM(amount)) OVER(ORDER BY date))  * 100 AS `growth`

FROM orders
GROUP BY month
ORDER BY date ASC
```

<div style="text-align:justify; color:black; font-size:20px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">What is the average insurance claimed by patients based on the number of children they have? Dont use Groupby</div><br>


```python
%%sql

SELECT children,claim,avg FROM 
    (SELECT *,
    AVG(claim) OVER(PARTITION BY children) AS avg,
    ROW_NUMBER() OVER(PARTITION BY children) AS row_num
    FROM insurance) t1

WHERE t1.row_num = 1
```

<br><br>
<div style="text-align:justify; color:black; font-size:20px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">What is the highest and lowest claimed amount by patients in each region? Dont use Groupby</div><br>



```python
%%sql
SELECT * FROM (

    SELECT region,
    
    FIRST_VALUE(claim) OVER(
        PARTITION BY region 
        ORDER BY claim DESC 
        ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING
        ) AS `highest_claim`,
    
    LAST_VALUE(claim) OVER(
        PARTITION BY region 
        ORDER BY claim DESC 
        ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING
        ) AS `lowest_claim`,
    
    ROW_NUMBER() OVER(PARTITION BY region) AS `row`

    FROM insurance
) t1

WHERE t1.row = 1
```

<br><br>
<div style="text-align:justify; color:black; font-size:20px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">Use MIN MAX function</div><br>



```python
%%sql
SELECT * FROM (

    SELECT region,
    
    MAX(claim) OVER(PARTITION BY region) AS `highest_claim`,
    MIN(claim) OVER(PARTITION BY region) AS `lowest_claim`,
    
    ROW_NUMBER() OVER(PARTITION BY region) AS `row`

    FROM insurance
) t1

WHERE t1.row = 1
```

<div style="text-align:justify; color:black; font-size:20px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">Identify the value clusters using RANGE mode with a window size of 1 preceding and 1 following.</div><br>



```python
%%sql
SELECT
    ID,
    Value,
    LAG(Value, 1) OVER (ORDER BY ID RANGE BETWEEN 1 PRECEDING AND 1 PRECEDING) AS prev_value,
    LEAD(Value, 1) OVER (ORDER BY ID RANGE BETWEEN 1 FOLLOWING AND 1 FOLLOWING) AS next_value
FROM sample_data;
```
