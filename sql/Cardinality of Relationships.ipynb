<h1 style="text-align:center; color:#005bbd; font-size:50px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px Black;">Cardinality of Relationships</h1>

<h2 style="text-align:justify; color:black; font-size:30px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; line-height:150%;"><p>Cardinality in database relationships refers to the number of occurrences of an
entity in a relationship with another entity.<br>Cardinality defines the number of
instances of one entity that can be associated with a single instance of the related
entity.</p></h2>


<h2 style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; line-height:150%;"><p>Entity Example</p>
