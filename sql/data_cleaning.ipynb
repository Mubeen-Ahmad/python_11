<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">Extract Info</div>



```python
%%sql 
SELECT Memory,

(CASE 
    WHEN Memory LIKE '%SSD%' AND Memory LIKE '%HDD%' THEN 'Hybrid (SSD + HDD)'
    WHEN Memory LIKE '%SSD%' THEN 'SSD'
    WHEN Memory LIKE '%Flash Storage%' THEN 'Flash Storage'
    WHEN Memory LIKE '%HDD%' THEN 'HDD'
    WHEN Memory LIKE '%Flash Storage%' AND Memory LIKE '%HDD%' THEN 'Hybrid(Flash + HDD)'
    WHEN Memory LIKE '%Hybrid%' THEN 'Hybrid'
    ELSE NULL
END) AS memory_type,

SUBSTRING_INDEX(Memory," ",1) AS primary_storage,

(CASE WHEN Memory LIKE '%+%' THEN SUBSTRING_INDEX(SUBSTRING_INDEX(Memory," ",-2)," ",1) ELSE 0 END) AS secondary_storage

FROM laptop_bk
LIMIT 40,10
```

<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">Add Columns and fill value</div>



```python
%%sql
ALTER TABLE laptop_bk

ADD COLUMN memory_type VARCHAR(100) AFTER Memory,
ADD COLUMN primary_storage VARCHAR(100) AFTER memory_type,
ADD COLUMN secondary_storage VARCHAR(100) AFTER primary_storage
```

```python
%%sql
UPDATE laptop_bk t1
SET memory_type = ( SELECT (CASE 
    WHEN Memory LIKE '%SSD%' AND Memory LIKE '%HDD%' THEN 'Hybrid (SSD + HDD)'
    WHEN Memory LIKE '%SSD%' THEN 'SSD'
    WHEN Memory LIKE '%Flash Storage%' THEN 'Flash Storage'
    WHEN Memory LIKE '%HDD%' THEN 'HDD'
    WHEN Memory LIKE '%Flash Storage%' AND Memory LIKE '%HDD%' THEN 'Hybrid(Flash + HDD)'
    WHEN Memory LIKE '%Hybrid%' THEN 'Hybrid'
    ELSE NULL
END) FROM laptop_bk t2 WHERE t2.index = t1.index
),

primary_storage = (SELECT SUBSTRING_INDEX(Memory," ",1) FROM laptop_bk t2 WHERE t2.index = t1.index),

secondary_storage = (SELECT (CASE 
                            WHEN Memory LIKE '%+%' THEN SUBSTRING_INDEX(SUBSTRING_INDEX(Memory," ",-2)," ",1) 
                            ELSE 0 
                            END) 
                    FROM laptop_bk t2 WHERE t2.index = t1.index
                    )
```

```python
%%sql
SELECT primary_storage,secondary_storage, 

(CASE 
 WHEN primary_storage LIKE "%TB%" THEN REGEXP_SUBSTR(primary_storage,'[0-9]+') * 1024 
 ELSE REGEXP_SUBSTR(primary_storage,'[0-9]+') END
 ) AS pri_GB,

(CASE 
 WHEN secondary_storage LIKE "%TB%" THEN REGEXP_SUBSTR(secondary_storage,'[0-9]+') * 1024 
 ELSE REGEXP_SUBSTR(secondary_storage,'[0-9]+') END
 ) AS sec_GB


FROM laptop_bk
LIMIT 40,10
```
