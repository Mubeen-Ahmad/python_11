<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;"><li>Check NULL Values</li></div>

```python
%%sql
SELECT 

SUM(id) IS NULL AS id,
SUM(Company) IS NULL AS Company,
SUM(TypeName) IS NULL AS TypeName,
SUM(Inches) IS NULL AS Inches,
SUM(width) IS NULL AS width,
SUM(height) IS NULL AS height,
SUM(touchscreen) IS NULL AS touchscreen,
SUM(ips_panel) IS NULL AS ips_panel,
SUM(cpu_name) IS NULL AS cpu_name,
SUM(cpu_speed) IS NULL AS cpu_speed,
SUM(Ram) IS NULL AS Ram,
SUM(memory_type) IS NULL AS memory_type,
SUM(primary_storage) IS NULL AS primary_storage,
SUM(secondary_storage) IS NULL AS secondary_storage,
SUM(gpu_brand) IS NULL AS gpu_brand,
SUM(gpu_name) IS NULL AS gpu_name,
SUM(OpSys) IS NULL AS OpSys,
SUM(Weight) IS NULL AS Weight,
SUM(Price) IS NULL AS Price

FROM laptop_bk
```

<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;"><li>Price</li></div>

```python
%%sql 

SELECT 

MIN(Price) AS min_price,
MAX(Price) AS max_price,
AVG(Price) AS avg_price,
STD(Price) AS std_price
FROM laptop_bk
```

<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;"><li>Find Quantiles of Price</li></div>

```python
%%sql

SELECT Price,percentile,Quartile FROM (

SELECT * FROM (

SELECT *,
( (count-1) / total) * 100 AS percentile,

    CASE 
        WHEN ((count-1) / total) * 100 >= 0 AND ((count-1) / total) * 100 < 25 THEN 'Q1 25%'
        WHEN ((count-1) / total) * 100 >= 25 AND ((count-1) / total) * 100 < 50 THEN 'Q2 50%'
        WHEN ((count-1) / total) * 100 >= 50 AND ((count-1) / total) * 100 < 75 THEN 'Q3 75%'
        WHEN ((count-1) / total) * 100 >= 75 AND ((count-1) / total) * 100 <= 100 THEN 'Q4 100%'
        ELSE NULL
    END AS Quartile

FROM 

(   
    SELECT *,
    COUNT(Price) OVER(ORDER BY Price ASC) AS count,
    COUNT(*) OVER() AS total
    FROM laptop_bk
    
) t1

) main

GROUP BY Quartile,percentile
ORDER BY percentile ASC
) t1

GROUP BY Quartile
```

<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;"><li>Find Outliers</li></div>

```python
%%sql

SELECT Price,percentile,Quartile FROM (

SELECT * FROM (

SELECT *,
( (count-1) / total) * 100 AS percentile,

    CASE 
        WHEN ((count-1) / total) * 100 >= 0 AND ((count-1) / total) * 100 < 25 THEN 'Q1 25%'
        WHEN ((count-1) / total) * 100 >= 25 AND ((count-1) / total) * 100 < 50 THEN 'Q2 50%'
        WHEN ((count-1) / total) * 100 >= 50 AND ((count-1) / total) * 100 < 75 THEN 'Q3 75%'
        WHEN ((count-1) / total) * 100 >= 75 AND ((count-1) / total) * 100 <= 100 THEN 'Q4 100%'
        ELSE NULL
    END AS Quartile

FROM 

(   
    SELECT *,
    COUNT(Price) OVER(ORDER BY Price ASC) AS count,
    COUNT(*) OVER() AS total
    FROM laptop_bk
    
) t1

) main

GROUP BY Quartile,percentile
ORDER BY percentile DESC
) t1

GROUP BY Quartile
```
