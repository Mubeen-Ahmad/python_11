<h1 style="text-align:center; color:#005bbd; font-size:50px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px Black;">Window Function</h1>


<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">Window functions in SQL are a type of analytical function that perform calculations
across a set of rows that are related to the current row, called a "window". <br>A
window function calculates a value for each row in the result set based on a subset
of the rows that are defined by a window specification.</div>



<br><br>
<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">The window specification is defined using the OVER() clause in SQL, which specifies
the partitioning and ordering of the rows that the window function will operate
on. <br><br>The partitioning divides the rows into groups based on a specific column or
expression, while the ordering defines the order in which the rows are processed
within each group.</div><br>

<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;"><br> Here i need a Avg rating of release_year SO i perform on group by on release_year</div><br>





```python
%%sql
SELECT *,AVG(rating) AS `year_avg` FROM data
GROUP BY release_year
```

<br><br>
<div style="text-align:justify; color:black; font-size:25px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">But here we need all Data With Avg? <br> SO i use Join also with subquery</div><br>





```python
%%sql
SELECT t1.id,t1.release_year,t1.rating,t2.year_avg FROM data t1
INNER JOIN (SELECT AVG(rating) AS `year_avg`,release_year FROM data GROUP BY release_year) AS t2
ON t1.release_year = t2.release_year
```

<h3 style="color:#005bbd; font-size:25px;">Aggregate AVG Function with OVER
</h3>
<h3 style="color:red; font-size:25px;">AVG(col) OVER(PARTITION BY column)</h3>

```python
%%sql
SELECT *,  

AVG(rating) OVER(PARTITION BY release_year) AS `year_avg`

FROM data
```

<div style="text-align:justify; color:black; font-size:28px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">Find top 2 most paying customers of each month
</div><br>






```python
%%sql
SELECT *,
MONTHNAME(date) AS 'month',
SUM(amount) AS 'total',

RANK() OVER(PARTITION BY 'month' ORDER BY total DESC) AS 'rank'
FROM orders

GROUP BY month,user_id
ORDER BY month DESC
```

<br><br>
<div style="text-align:justify; color:black; font-size:28px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black;">Now need top 2 most paying customers of each month<br><br> SO use Subquery and use where
</div><br>






```python
%%sql
SELECT * FROM (
                SELECT user_id,
                MONTHNAME(date) AS `month`,
                SUM(amount) AS 'total',
                RANK() OVER(PARTITION BY `month` ORDER BY total DESC) AS `rank`
                FROM orders
                GROUP BY month,user_id
                ORDER BY month DESC
              ) t1


WHERE t1.rank < 3
ORDER BY month DESC, rank ASC
```

<h3 style="color:#005bbd; font-size:25px;">Aggregate LAG() Function with OVER
</h3>
<h3 style="color:red; font-size:25px;">LAG(column) OVER(PARTITION BY col ORDER BY col DESC)</h3><br>
The Lag function in MySQL is a window function that provides you with the value of a column in the previous row.</div>





```python
%%sql
SELECT *,LAG(marks) OVER(ORDER BY student_id DESC)
FROM marks
```
