# Sigmoid Function

- ### The sigmoid function is a mathematical function that maps any input value to a value between 0 and 1. It is a commonly used activation function in machine learning and deep learning algorithms, particularly in binary classification tasks.
<br>

# Formula
<br>

- ## Sigmoid(x) =  1/ 1+e ^ -x

```python
import numpy as np
```

```python
def sigmoid(a):
    return 1 / (1 + np.exp( -(a) ))
```

```python
arr = np.arange(100)
arr
```

```python
sigmoid(arr)
```

# Mean Square Error 

- ### The mean square error (MSE) is a commonly used metric for measuring the difference between an observed value and the predicted value of a model. It is calculated as the average of the squared differences between the predicted values and the actual values.
<br>

# Formula
<br>

- ## MSE = mean( (actual - predicted) ** 2 )

```python
def MSE(actual,pred):
    return np.mean( (actual-pred) ** 2 )
```

```python
actual = np.random.randint(1,50,25)
actual
```

```python
pred = np.random.randint(1,50,25)
pred
```

```python
MSE(actual,pred)
```

#  Binary cross-entropy

- ### Binary cross-entropy, also known as log loss, is a loss function commonly used in binary classification tasks. It is a measure of the dissimilarity between the predicted probability distribution and the true probability distribution of binary labels.
<br>

# Formula
<br>

- ## bce_loss = - (y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

```python
def BCE(actual,pred):
    return -(actual * np.log(pred) + (1 - actual) * np.log(1 - pred))
```

```python
actual = 1
pred = 0.8
```

```python
BCE(actual,pred)
```
