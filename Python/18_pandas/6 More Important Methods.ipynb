### FIND the last match played by virat kohli in Delhi

```python
ipl = pd.read_csv("./datasets/ipl-matches.csv")
temp = ipl[["Team1Players","Team2Players","City","Date"]].copy()
temp.head()
```

```python
temp["all_players"] = temp["Team1Players"] + temp["Team2Players"]

temp.head()
```

```python
mask = temp["all_players"].apply(lambda x: "V Kohli" in x)
mask
```

```python
temp["vk played"] = mask
temp.head()
```

```python
extract = temp[(temp["City"] == "Delhi") & (temp["vk played"] == True)].sort_values("Date",ascending=False).copy()
extract
```

```python
extract.drop_duplicates(subset=["City","vk played"],keep="first")
```

```python
# apply on dataframe

points = {
    "1st_p" : [(3,4),(-6,5),(0,0),(-10,1),(4,5)],
    "2nd_p" : [(-3,4),(0,0),(2,2),(10,10),(1,1)]  
}

df_points = pd.DataFrame(points)
df_points
```

```python
def euclidean(row):
    
  a = row["1st_p"]
  b = row["2nd_p"]
    
  return ( ((a[1] - a[0]) ** 2) + ((b[1]  - b[0]) ** 2) ) ** 0.5
```

```python
df_points.apply(euclidean,axis=1)
```

```python
df_points["distance"] = df_points.apply(euclidean,axis=1)
df_points
```
