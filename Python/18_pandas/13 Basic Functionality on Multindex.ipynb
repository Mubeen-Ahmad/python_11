```python
import pandas as pd
import numpy as np
```

- ### MultiIndex DF Support all funcationality of normal DF
- ### Indexing and Slicing
- ### Index Sorting
- ### Transpose
- ### Swaping


```python
col = pd.MultiIndex.from_product([
    ["Lahore","Karachi"],
    ["avg_pkg","std"]
    ])
col
```

```python
ind = pd.MultiIndex.from_product(
    [ 
        ["cse","ece"], [2019,2020,2021,2022] 
     
     ])

ind
```

```python
# add in df
num = np.arange(4*8).reshape(8,4)
mul_df = pd.DataFrame(num,columns=col,index=ind)
mul_df
```
