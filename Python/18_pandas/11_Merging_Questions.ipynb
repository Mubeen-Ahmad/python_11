### Q-6 From the IPL wala dataset you have to find the Purple cap holder each season.
<br>

### *Note: Bowler with most no wickets in a season gets purple cap. If more than one bowler have same no of wickets in the season, one with least ecomnomy among them is purple cap holder.*
<br>

### Bowler's Economy = runs-conceded per six balls

```python
balls = pd.read_csv("./datasets/IPL_Ball_by_Ball_2008_2022.csv")
balls.head()
```

```python
matches = pd.read_csv("./datasets/IPL_Matches_2008_2022.csv")
```

```python
matches.head()
```

```python
season = balls.merge(matches[["ID","Season"]],how="inner",on="ID")
season.head()
```

```python
season["IsBowlerWicket"] = season["kind"].apply(lambda x: 1 if x in ["caught", 'caught and bowled', 'bowled', 'stumped','lbw', 'hit wicket'] else 0)
```

```python
season.IsBowlerWicket
```

```python
season["BowlerRun"] = season["extra_type"].apply(lambda x: 0 if x in ["legbyes", "byes"] else 1) * season["total_run"]
```

```python
season.head()
```

```python
season["IsLegalBall"] = season["extra_type"].apply(lambda x: 0 if x in ["wides", "noballs"] else 1)
season.head()
```

```python
pcapdf = season.groupby(["Season", "bowler"], as_index = False)[["IsBowlerWicket", "BowlerRun", "IsLegalBall"]].sum()
pcapdf
```

```python
pcapdf["Economy"] = pcapdf["BowlerRun"] / pcapdf["IsLegalBall"] * 6
pcapdf
```

```python
final = pcapdf.sort_values(by=["Season","IsBowlerWicket","Economy"],ascending=[True,False,True])
final
```

```python
# drop duplicates Season
final.drop_duplicates(subset="Season",keep="first")
```
