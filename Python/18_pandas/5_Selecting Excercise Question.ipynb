## 7 Write a function that can return the track record of 2 teams against each other

```python
df_ipl[["Team1","Team2"]]
```

```python
team1 = "Kings XI Punjab"
team2 = "Mumbai Indians"
```

```python
def track(t1,t2):
    mask1 = (df_ipl.Team1 == t1) | (df_ipl.Team2 == t1) 
    mask2 = (df_ipl.Team1 == t2) | (df_ipl.Team2 == t2)
    tmp_data = df_ipl[mask1 & mask2].copy()
    total_match = tmp_data.shape[0]
    
    print("Total Match",total_match)
    
    if len(tmp_data["WinningTeam"].unique()) == 1:
        if tmp_data["WinningTeam"].unique()[0] == t1:
            print(t1," Wins",total_match,"Times")
            print(t2," Wins 0 Times")
        else:
            print(t2," Wins",total_match,"Times")
            print(t1," Wins 0 Times")
    else:
        tmp = tmp_data["WinningTeam"].value_counts()
        print(tmp.keys()[0],"Wins",tmp.values[0],"Times")
        print(tmp.keys()[1],"Wins",tmp.values[1],"Times")
```

```python
track(team1,team2)
```

```python
track("Sunrisers Hyderabad","Mumbai Indians")
```
