### Q-1: You are given a Multi index DataFrame. You task is to swap level-0 index with level-0 column.
<br>

- ### Change Branch -> ('cse', 'ece') as columns at level-0 and ('Karachi'	'Lahore') as level-0 Index
<br>

- ### And Sort on row index level-0

```python
import pandas as pd, numpy as np

# Given Code Snippets
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
df = pd.DataFrame(
    [
        [1,2,0,0],
        [3,4,0,0],
        [5,6,0,0],
        [7,8,0,0],
        [9,10,0,0],
        [11,12,0,0],
        [13,14,0,0],
        [15,16,0,0],
    ],
    index = multiindex,
    columns = pd.MultiIndex.from_product([['karachi','lahore'],['avg_package','students']])
)

df
```

```python
df.unstack(level=0).stack(level=0).swaplevel().swaplevel(axis=1).sort_index().sort_index(axis=1)
```

### Q-2: Covid Cases Data Set Problem
 <br>

- ### Make a DataFrame Using both Covid Dataset. 
<br>

- ### With Country name as Level-0 and Provinance/State as Level-1 Index.
<br>

- ### Date No of cases and No of Deaths as Columns.

```python
death = pd.read_csv("./datasets/covid_confirm.csv")
confirm = pd.read_csv("./datasets/covid_deaths.csv")
```

```python
confirm.head()
```

```python
death.head()
```

```python
confirm.head()
```

```python
dth = death.melt(id_vars=["Province/State","Country/Region","Lat","Long"],var_name="date",value_name="Deaths")

cnf = confirm.melt(id_vars=["Province/State","Country/Region","Lat","Long"],var_name="date",value_name="Confirm Cases")
```

```python
dth.head(2)
```

```python
cnf.head(2)
```

```python
# now merge the data
final = cnf.merge(dth,how="inner",on=["Province/State","Country/Region","Lat","Long","date"])[["Country/Region","Province/State","date","Confirm Cases","Deaths"]]   

final
```

```python
final.set_index(["Country/Region","Province/State"],inplace=True)
```

```python
final
```

### Q-3: Show Country with Highest death percent out of confirmed Cases. 

```python
tmp = final.groupby("Country/Region").sum()
tmp["percentage"] = np.round((tmp["Deaths"] / tmp["Confirm Cases"]) * 100,1)
```

```python
tmp.sort_values(by="Deaths",ascending=False).drop(columns=["date"]).reset_index()
```

### Q-4 : Make a dataframe for Pakistan from Covid Data with one extra column representing no of new cases. 

- ###  Just for Assumption "No of new cases" will be equal to difference of "no of cases" with previous day.

<br>

- ### Note:- Try using shift Function

<br>

- ### get record only 2022 1st two months

```python
country = final.loc["Pakistan"].copy()
country
```

```python
# convert date
country["date"] = pd.to_datetime(country["date"].values,format='%m/%d/%y').strftime('%d-%m-%Y')
```

```python
# split year
country["year"] = country["date"].apply(lambda x :x.split("-")[2])
```

```python
# split month
country["month"] = country["date"].apply(lambda x :x.split("-")[1])
```

```python
# split date
country["Date"] = country["date"].apply(lambda x :x.split("-")[0])
```

```python
# difference previous case
country["new_cases"] = country["Confirm Cases"] - country["Confirm Cases"].shift(1)
```

```python
country
```

```python
# 1st month
country[(country["month"] == "01") & (country["year"] == "2022")].sort_values(by="Date")
```

```python
# 2nd month
country[(country["month"] == "02") & (country["year"] == "2022")].sort_values(by="Date")
```
