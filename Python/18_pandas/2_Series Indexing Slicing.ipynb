# So Soloution is use iloc

- ### Specifically, iloc stands for "integer location" and takes integer-based indices to select specific rows or columns.

# Boolean Indexing

```python
marks = [67,37,49,70]
subjects = ['maths','english','science','Urdu']

data = pd.Series(marks,index=subjects)
data
```

```python
data[data > 50]
```

```python
data[data < 50]
```

```python
data[(data > 50) & (data %2 != 0)]
```
