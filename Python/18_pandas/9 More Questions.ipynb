- ### 1 Group the data by 'Team' and 'batter', and calculate the total number of runs scored by each batter in each team.

```python
def joins(x):
    return "|".join(x.values)
    
    
join_teams = ipl_df[["BattingTeam","BowlingTeam"]].apply(joins,axis=1)

ipl_df["Batting|Bowling"] = join_teams
```

```python
ipl_df.sample()[["BattingTeam","BowlingTeam","Batting|Bowling"]]
```

```python
result = ipl_df.groupby("Batting|Bowling")[["total_run"]].sum().sort_values(by="total_run",ascending=False).reset_index()
```

```python
result[['BattingTeam', 'BowlingTeam']] = result["Batting|Bowling"].apply(lambda x: pd.Series(x.split("|")))
```

```python
result.drop(columns="Batting|Bowling")
```

- ### 3 Using the same Netflix titles dataset, group the data by rating and calculate the average duration (in minutes) for each rating. Sort the data in descending order of average duration.

```python
netflix.sample(2)
```

```python
netflix_df = netflix.dropna()
```

```python
def extract(x):
    return x["duration"].str.extract("(\d*)").astype(int).mean()
duration = netflix_df[["rating","duration"]].groupby("rating").apply(extract).sort_values(by=0)
```

- ### 9 Using the New York City Airbnb dataset, group the data by host ID and calculate the average rating and number of reviews for each host. 
<br>

- ### Sort the data in descending order of average rating and ascending order of number of reviews. 
<br>
    
- ### Also, split the name column into separate words and find the most common words in the names of the top 10 hosts with the highest average rating.

```python
new_york.sample(3)
```

```python
temp = new_york.dropna().copy()

temp["review_scores_rating"] = (temp['number_of_reviews'] * temp['reviews_per_month']) / temp['number_of_reviews']
temp.head()
```

```python
host_stats = temp.groupby('host_id').agg(
    {'review_scores_rating': 'mean', 
     'number_of_reviews': 'sum',
    'name':'first'}).sort_values(['review_scores_rating', 'number_of_reviews']
                                 , ascending=[False, True])

host_stats
```

```python
top_10 = host_stats.head(10)
top_10
```
