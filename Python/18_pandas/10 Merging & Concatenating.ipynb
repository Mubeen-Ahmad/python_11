# Merging 

- ### inner merging
    - ### Shows only columns key data means
    <br>
    
    - ### .merge(regs,how='inner',on="column name")
<br>

- ### left merging
    - ### merge according to Left df column(key)
    - ### if left column(key) not exit than return nan
    <br>
    
    - ### .merge(regs,how='left',on="column name")

<br>

- ### right merging
    <br>
    
    - ### Show Specfic data ; 
    - ### if Right column(key) not exit than return nan
    <br>
    
    - ### .merge(regs,how='right',on="column name")
 <br>
 
- ### outer merging
    - ### Shows All Data
    <br>
    
    - ### .merge(regs,how='outer',on="column name")
<br>

- ### self merging
    - ### Shows Matching A,B Data
    <br>
    
    - ### .merge(regs,how='inner',left_on,right_on)

```python
regs.tail()
```

```python
# here we have 2 course_id column 
# regs and couse


# consider course[course_id] is a  Left
# consider regs[course_id] is a Right

# consider .merge[course_id] is the Main

# Left course["course_id"] column 11,12 are exist 
# Right regs["course_id"] column 11,12 are not exist 

# So return Nan Value in Main 
```

```python
course["course_id"].unique()
```

```python
course.tail(2)
```

```python
regs.tail(2)
```

- ### right merging

```python
tmp = pd.DataFrame({
    "student_id":[26,27,28],
    "name":["Rizwan","Anon","Imran"],
    "partner":[28,26,27]
})
tmp
```

```python
students = pd.concat([students,tmp],ignore_index=True)
students
```

```python
students.tail(3)
```

```python
students.merge(regs,how="right",on="student_id")
```

```python
students.tail(3)
```

```python
regs.tail(3)
```

```python
# here we have 2 course_id column 
# students[student_id] and regs[student_id]

# consider regs[student_id] is a  left
# consider students[student_id] is a right

# consider .merge[student_id] is the main

# IN left column   regs["student_id"] [42,50,38] are include
# BUT Right column students["student_id"] [42,50,38] are not exits

# So return NaN value
students.merge(regs,how="right",on="student_id").tail(3)
```

- ### outer merging

```python
regs.tail(8)
```

```python
students.tail(3)
```

```python
# in left join
students.merge(regs,how="left",on="student_id").tail(8)
```

```python
# in right join
students.merge(regs,how="right",on="student_id").tail(8)
```

```python
# In outer joining the columns on both left and right
# IF the keys(column value) are missing than they fill with nan
students.merge(regs,how="outer",on="student_id").tail(8)
```

## Find top 3 Stadiums with highest sixes/match ratio

```python
matches = pd.read_csv("./datasets/matches.csv")
matches.head(3)
```

```python
deliveries = pd.read_csv("./datasets/deliveries.csv")
deliveries.head(3)
```

```python
df = deliveries.merge(matches,how="inner",left_on="match_id",right_on="id")
df
```

```python
six =df["batsman_runs"] == 6
six
```

```python
six_df = df[six]
six_df
```

```python
no_of_six = six_df.groupby("venue")["venue"].count()
no_of_six
```

```python
df.head()
```

```python
no_of_matched = matches.venue.value_counts()
no_of_matched
```

```python
(no_of_six / no_of_matched).sort_values(ascending=False).head(5)
```

## Find orange cap holder of all season

```python
matches.head()
```

```python
deliveries.head()
```

```python
selected_1 = matches[["id","season"]]
selected_1
```

```python
selected_2 = deliveries[["match_id","batsman","batsman_runs"]]
selected_2
```

```python
mrg = selected_1.merge(selected_2,how="inner",left_on="id",right_on="match_id")
mrg
```

```python
srt = mrg.groupby(["season","batsman"])["batsman_runs"].sum().reset_index()
srt
```

```python
srt = srt.sort_values(by="batsman_runs",ascending=False)
srt
```

```python
srt.drop_duplicates(subset=["season"],keep="first")
```

## we also use groupby().first()

```python

srt.groupby("season").first().sort_values(by="batsman_runs",ascending=False).reset_index()
```
