# What is groupby
<br>

- ### pandas groupby is a powerful function for grouping data based on one or more columns of a DataFrame. 
<br>

- ### It allows you to split the DataFrame into groups based on the values in one or more columns, and then perform operations on each group.
<br>

- ### When using groupby, you first select the columns that you want to group by, and then apply a function to each group. 
<br>

- ### The resulting output is a DataFrame with the same columns as the original DataFrame, but where each row represents a group and the values are the result of the function applied to that group.
